```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

# File path
vix_csv = "vix.csv"

# Read VIX data file and store into Pandas DataFrames
vix_df = pd.read_csv(vix_csv)

vix_df.head()
```

```python
vix_df["DATE"] = pd.to_datetime(vix_df["DATE"])

vix_df.set_index('DATE', inplace=True)

vix_df.head()
```

```python
vix_df_mean = vix_df.resample('MS').mean()

vix_df_mean = vix_df_mean[["CLOSE"]]

vix_df_mean.head()
```

```python
vix_df_mean.columns
```

```python
vix_df_mean = vix_df_mean.reset_index()

vix_df_mean.head()
```

```python
vix_df_mean.columns
```

```python
# File path
music_csv = "clean_music_file.csv"

music_df = pd.read_csv(music_csv)

music_df.head()
```

```python
music_df["date"] = pd.to_datetime(music_df["date"])

music_df.set_index('date', inplace=True)

music_df.head()
```

```python
music_df_sum = music_df.resample('MS').sum()

music_df_sum.head()
```

```python
music_mood_df = music_df_sum[["valence"]]

music_mood_df.head()
```

```python
music_mood_df = music_mood_df.reset_index()

music_mood_df.head()
```

```python
vix_music_merge = pd.merge(vix_df_mean, music_mood_df, how="outer", on="DATE")

vix_music_merge.head()
```

```python
figure1 = vix_df_mean["CLOSE"].plot(kind="line", figsize=(8,6),
                                title="VIX Daily Close",
                                xlabel="Date",
                                ylabel="Close")
```

```python
figure3 = music_mood_df["valence"].plot(kind="line", figsize=(8,6),
                                title="Music Valence",
                                ylabel="Songs",ylim=(2000,3000))

```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(vix_df_mean.DATE,
        vix_df_mean.CLOSE,
        color="red", 
        marker="o")
# set x-axis label
ax.set_xlabel("Month", fontsize = 14)
# set y-axis label
ax.set_ylabel("Close",
              color="red",
              fontsize=14)
```
